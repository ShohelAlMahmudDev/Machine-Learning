**Resturant Data(tips)**

```python
from  sklearn.linear_model import LinearRegression
# Load Dataset from Seaborn
df = sns.load_dataset('tips')
df.head()
```

### Steps
**1. Select the Feature Column from dataframe**

```python
#double column value because it is depends on another column
x = df[['total_bill']]
```

**2 Select the Target(Label) Column from dataframe**

```python
y = df['tip']
```

### check the plot how its look like 

```python
sns.lmplot(x='total_bill',y='tip', data=df)
```

```python
#Initialize the model
model=LinearRegression()
#trained the model
model.fit(x, y)
```

### Lets Predict the tip for unknown total bill

```python
model.predict([[50]])
```

## Complete Code Together

```python
import pandas as pd
import seaborn as sns
from  sklearn.linear_model import LinearRegression

import warnings

df = sns.load_dataset('tips')
df.head()
#double column value because it is depends on another column
#select the feature column from dataframe(df)
X_bill = df[['total_bill']]
#select the target(label) column from dataframe(df)
y_tip = df['tip']
#Initialize the model
model=LinearRegression()
#trained the model
model.fit(X_bill, y_tip)
#Predict the tip

warnings.filterwarnings("ignore", category=UserWarning)
X_test=[[50]]
model.predict(X_test)
```
